```python
import pytesseract
from pdf2image import convert_from_path
import tempfile
import os

# Path to the PDF file
pytesseract.pytesseract.tesseract_cmd = 'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
def extract_text(pdf_path):
    # Create a temporary directory to store images
    with tempfile.TemporaryDirectory() as path:
        # Convert each page of the PDF to an image
        images = convert_from_path(pdf_path, output_folder=path,first_page=1, last_page=1, fmt='png')
    
        # Initialize an empty string to store extracted text
        extracted_text = ''
    
        # Iterate through each image (page) and extract text
        for i, image in enumerate(images):
            # Use pytesseract to extract text from the image
            extracted_text = pytesseract.image_to_string(image)
    
            # Append the extracted text to the final result
            extracted_text += f'Page {i + 1}:\n{extracted_text}\n\n'
    
    # Print or save the extracted text
    return extracted_text
```

```python

#extract re
import re

def extract_subject(extracted_text):
    subject = None
    pattern = r'RE:\s*(.*(?:\n.+)+)'
    match = re.search(pattern, extracted_text, re.IGNORECASE | re.DOTALL)
    if match:
        subject = match.group(1).strip().split('\n')[0]+match.group(1).strip().split('\n')[1]
         
              
    return subject
```

```python
# rename folder
def rename_pdf(pdf_path, new_name):
    # Rename the PDF file
    directory, filename = os.path.split(pdf_path)
    new_path = os.path.join(directory,  new_name)
    os.rename(pdf_path, new_path)
```

```python
pdf_directory = "C:\\Users\\user\Downloads\\OneDrive_2_3-7-2024"

# Iterate over all PDF files in the directory
for filename in os.listdir(pdf_directory):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(pdf_directory, filename)
        extracted_text = extract_text(pdf_path)
        if extracted_text:
            subject = extract_subject(extracted_text)
            if subject:
                new_name = f"{subject}.pdf"
                # Replace or remove special characters from the filename
                cleaned_filename = ''.join(c for c in new_name if c.isalnum() or c in [' ', '_', '-', '.'])
                cleaned_pdf_path = os.path.join(pdf_directory, cleaned_filename)
                counter = 1
                while os.path.exists(cleaned_pdf_path):
                    # If the file already exists, append a counter suffix to the filename
                    base, ext = os.path.splitext(cleaned_pdf_path)
                    cleaned_pdf_path = f"{base}_{counter}{ext}"
                    counter += 1
                rename_pdf(pdf_path, cleaned_pdf_path)
                print(f"Renamed {filename} to {os.path.basename(cleaned_pdf_path)}")
            else:
                print(f"Subject not found in {filename}, skipping...")
        else:
            print(f"Unable to extract text from {filename}, skipping...")
```
